```python
sorted_df= df.sort_values(by=['horsepower'])
sorted_df
```

```python
plt.plot(sorted_df['horsepower'],sorted_df['acceleration'])
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.show()
weight_sorted=df.sort_values(by=['weight'])
plt.plot(weight_sorted['weight'],weight_sorted['acceleration'])
plt.xlabel('Weight')
plt.ylabel('Acceleration')
plt.show()
plt.plot(weight_sorted['weight'],weight_sorted['horsepower'])
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.show()
```

# Learning outcomes from above three plots

## HP V/S Acc.
### From this plot we come to know that acceleration of a car decreases as horsepower increases, which is quite strange as we expect acc. to increase with HP. To find out reason for this strange behaviour we plot
## Weight V/S Acc.
### From this plot we come to know that with increase in weight acc. decreases, which is expected. But by seeing the nature of graph it seems that the effect is quite neglible.
## Weight V/S HP
### This graph shows exactly what anyone would expect! It shows that as hp increases weight also increases as we have to make engine components beefier to handle the power or we can also say that if your vehicle is massive you need more hps to move it with a decent speed
